```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the dataset (assuming the file is named 'student_performance.csv')
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Assignment/Student Performance/Raw Data/StudentsPerformance.csv')

# Display the first few rows to confirm data loading
print("Raw Data:\n", df.head())

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Save cleaned data to a 'cleaned' folder
if not os.path.exists('cleaned'):
    os.makedirs('cleaned')
cleaned_file_path = '/content/drive/MyDrive/Colab Notebooks/Assignment/Student Performance/Data Cleaned/student_performance_cleaned.csv'
df.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to: {cleaned_file_path}")

# Summary statistics for numerical columns (math, reading, and writing scores)
summary_stats = df[['math score', 'reading score', 'writing score']].describe()
print("\nSummary Statistics:\n", summary_stats)

# Save the summary statistics to a text file
with open('/content/drive/MyDrive/Colab Notebooks/Assignment/Student Performance/Analysis/student_performance_summary.txt', 'w') as f:
    f.write("Summary Statistics:\n")
    f.write(summary_stats.to_string())

# Analysis and Visualization

# 1. Gender vs Average Scores (Math, Reading, Writing)
df['Average Score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='Average Score', data=df)
plt.title('Average Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.grid(True)
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Assignment/Student Performance/Analysis/average_score_by_gender.png')
plt.show()

# 2. Parental Level of Education vs Average Scores
plt.figure(figsize=(10, 6))
sns.barplot(x='parental level of education', y='Average Score', data=df)
plt.title('Average Score by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Assignment/Student Performance/Analysis/average_score_by_parental_education.png')
plt.show()

# 3. Test Preparation Course vs Scores
plt.figure(figsize=(8, 6))
sns.boxplot(x='test preparation course', y='Average Score', data=df)
plt.title('Average Score by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Average Score')
plt.grid(True)
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Assignment/Student Performance/Analysis/average_score_by_test_preparation.png')
plt.show()

# 4. Math Score Distribution by Lunch Type
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='math score', hue='lunch', multiple="stack", bins=10)
plt.title('Math Score Distribution by Lunch Type')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Assignment/Student Performance/Analysis/math_score_distribution_by_lunch.png')
plt.show()

print("Analysis Complete. Plots and summary saved.")
```
