```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create 'cleaned' folder if it doesn't exist
if not os.path.exists('cleaned'):
    os.makedirs('cleaned')

# Load the raw data from CSV file
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Assignment/Frality/Raw Data/sample_data.csv')

# Display the first few rows to confirm data loading
print("Raw Data:\n", df)

# Perform any data cleaning steps if necessary (for now, assume data is already clean)
# For example, handling missing values, but in this case, data seems clean

# Save the cleaned data to a 'cleaned' folder
cleaned_file_path = '/content/drive/MyDrive/Colab Notebooks/Assignment/Frality/Data Cleaned/sample_cleaned.csv'
df.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned data saved to: {cleaned_file_path}")

# Summary statistics for height, weight, age, and grip strength
summary_stats = df.describe()
print("\nSummary Statistics:\n", summary_stats)

# Create analysis plots
# Plot 1: Scatter plot of Age vs Grip Strength with Frailty indicated by color
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Grip Strength', hue='Frailty', data=df, palette='coolwarm')
plt.title('Age vs Grip Strength (Colored by Frailty)')
plt.xlabel('Age (years)')
plt.ylabel('Grip Strength (kg)')
plt.grid(True)
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Assignment/Frality/Analysis/age_vs_grip_strength.png')
plt.show()

# Plot 2: Box plot of Grip Strength based on Frailty status
plt.figure(figsize=(8, 6))
sns.boxplot(x='Frailty', y='Grip Strength', data=df)
plt.title('Grip Strength by Frailty Status')
plt.xlabel('Frailty')
plt.ylabel('Grip Strength (kg)')
plt.grid(True)
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Assignment/Frality/Analysis/grip_strength_by_frailty.png')
plt.show()

# Plot 3: Histogram of Grip Strength
plt.figure(figsize=(8, 6))
plt.hist(df['Grip Strength'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Grip Strength')
plt.xlabel('Grip Strength (kg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Assignment/Frality/Analysis/grip_strength_histogram.png')
plt.show()

# Save the summary statistics to a text file
with open('/content/drive/MyDrive/Colab Notebooks/Assignment/Frality/Analysis/frailty_analysis_summary.txt', 'w') as f:
    f.write("Summary Statistics:\n")
    f.write(summary_stats.to_string())

print("Analysis Complete. Plots, summary, and cleaned data saved.")
```
